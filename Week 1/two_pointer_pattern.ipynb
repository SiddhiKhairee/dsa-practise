### Introduction

Two-pointers is a problem solving pattern.
We can use it to traverse a data-structure such that the pointers can be moved towards each other or away frm each other.

Used to reduce the time complexity. 

Pointer: A variable that represents the index or position in a data structure.




Using these pointers we can compare the elements within the data structures without having to rely to nested loops. 

The following are the three most common type of problems where we use two pointers: 

### CONVERGING POINTERS

- Pointers start at the opposite ends of the data structure
- As we iterate, the pointers adjust their positions such that they move towards each other. 
- This movement is based on comparisons, until certain condition is met or they cross each other. 
- This is useful for the problems where we need to check the elements on the opposite ends of the data structure.

For example: Check whether a string is a palindrome or not. 

Step 1: Intialize two pointer at the end of the string.

Step 2: Compare their element, if they match, move inwards

Step 3: If they don't match return false

Step 4: Keep repeating, until pointers cross each other or end up on the same variable. 


### Parallel pointers

- In this approach, both the pointer start at the same end, and they move in the same direction. 
- In this the right pointer, is used to find new information 
- The left pointer is used to track progress or maintain constraints. 
- For example: The sliding window technique. 
- In this technique two pointers, slide accross an array or string while maintaing a dynamic range. 


###  Trigger Based Pointers

- Both pointers move independently.
- We keep on moving the first pointer independently until it comes accross an element that meets some condition.
- After the condtion is met, we start the traversing of the second pointer to find additional information, related to what the first pointer has found. 

For example: Find the nth (N=4) node from the end of the linkedlist.

Step 1: Move the first pointer 4 steps forward. 

Step 2: Then as the first pointer is moving, consecutively move the second pointer along with that.

Step 3: When our first pointer hits the end of the linked list, our first pointer will be at (N-4)th element. 




### When to use Two-Pointers

- Mainly used for linear data structures such as arrays, strings, or linked lists. 
- The next clue where two pointers would be useful is when the linear data structure's input follows a certain patter. For instance: a sorted array or a palindrome.

### Leetcode Example 1: Two sum Problem on a sorted array

Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.

Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

Your solution must use only constant extra space.

 

Example 1:

Input: numbers = [2,7,11,15], target = 9
Output: [1,2]
Explanation: The sum of 2 and 7 is 9. Therefore, index1 = 1, index2 = 2. We return [1, 2].
Example 2:

Input: numbers = [2,3,4], target = 6
Output: [1,3]
Explanation: The sum of 2 and 4 is 6. Therefore index1 = 1, index2 = 3. We return [1, 3].
Example 3:

Input: numbers = [-1,0], target = -1
Output: [1,2]
Explanation: The sum of -1 and 0 is -1. Therefore index1 = 1, index2 = 2. We return [1, 2].
 

Constraints:

2 <= numbers.length <= 3 * 104
-1000 <= numbers[i] <= 1000
numbers is sorted in non-decreasing order.
-1000 <= target <= 1000
The tests are generated such that there is exactly one solution.

```python
from typing import List
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        output_indices = [] 
        # numbers = [2,7,11,15]

        left_pointer = 0
        right_pointer = len(numbers)-1

        while(left_pointer < right_pointer):

            current_sum = numbers[left_pointer] + numbers[right_pointer]

            if(current_sum == target):
                output_indices.append(left_pointer+1)
                output_indices.append(right_pointer+1)
                break
            
            elif(current_sum < target):
                left_pointer += 1

            else:
                right_pointer -= 1

        return output_indices

        
```

### Leetcode Example 2: Remove Duplicates from Sorted Array

Consider the number of unique elements in nums to be k​​​​​​​​​​​​​​. After removing duplicates, return the number of unique elements k.

The first k elements of nums should contain the unique numbers in sorted order. The remaining elements beyond index k - 1 can be ignored.

If all assertions pass, then your solution will be accepted.

 

Example 1:

Input: nums = [1,1,2]
Output: 2, nums = [1,2,_]
Explanation: Your function should return k = 2, with the first two elements of nums being 1 and 2 respectively.
It does not matter what you leave beyond the returned k (hence they are underscores).
Example 2:

Input: nums = [0,0,1,1,1,2,2,3,3,4]
Output: 5, nums = [0,1,2,3,4,_,_,_,_,_]
Explanation: Your function should return k = 5, with the first five elements of nums being 0, 1, 2, 3, and 4 respectively.
It does not matter what you leave beyond the returned k (hence they are underscores).
 

Constraints:

1 <= nums.length <= 3 * 104
-100 <= nums[i] <= 100
nums is sorted in non-decreasing order.

```python
nums = [0,0,1,1,1,2,2,3,3,4]

def remove_dup(arr):
    l = len(arr)
    # output_arr = []
    lp = 0
    rp = lp+1
    while(rp<len(arr)):
        if(arr[lp] == arr[rp]):
            arr.pop(rp)
        else:
            lp += 1
            rp += 1

    

    for i in range(len(arr), l):
        arr.append(0)


    return len(arr), arr

print(remove_dup(nums))
```

### Problem 3: Valid Palindrome
A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.

 

Example 1:

Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.
Example 2:

Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.
Example 3:

Input: s = " "
Output: true
Explanation: s is an empty string "" after removing non-alphanumeric characters.
Since an empty string reads the same forward and backward, it is a palindrome.
 

Constraints:

1 <= s.length <= 2 * 105
s consists only of printable ASCII characters.

```python
s = "a"
s_lower = s.lower()
print(s_lower)
```

```python
s_lower_without_space = s_lower.replace(" ", "")
print(s_lower_without_space)
```

```python
alphanumeric_chars = [char for char in s_lower if char.isalnum()]
print(alphanumeric_chars)
```

```python
from typing import List
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s_lower = s.lower()
        s_list = [i for i in s_lower if i.isalnum()]
        flag = 0
        lp = 0
        rp = len(s_list)-1

        while(lp<rp):
            if(s_list[lp] == s_list[rp]):
                lp += 1
                rp -= 1
            
            else: 
                flag = 1
                break
        
        if flag==0:
            return True
        else:
            return False
```

### Problem 4: Container with Most Water
You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

 

Example 1:


Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
Example 2:

Input: height = [1,1]
Output: 1
 

Constraints:

n == height.length
2 <= n <= 105
0 <= height[i] <= 104

```python
height = [1,8,6,2,5,4,8,3,7]


def water_container(h):
    lines = [i+1 for i in range(len(h))] 
    area_arr = []
    # initialize the pointers
    lp = 0
    rp = len(h) - 1 
    area = (lines[rp] - lines[lp]) * (min(h[lp], h[rp]))
    area_arr.append(area)
    while(lp<rp):

        if (h[lp] < h[rp]):
            lp += 1
            area = (lines[rp] - lines[lp]) * (min(h[lp], h[rp]))
            area_arr.append(area)
        
        elif(h[lp] > h[rp]):
            rp -= 1
            area = (lines[rp] - lines[lp]) * (min(h[lp], h[rp]))
            area_arr.append(area)

        else:
            lp += 1
            rp -= 1
            area = (lines[rp] - lines[lp]) * (min(h[lp], h[rp]))
            area_arr.append(area)

    return max(area_arr)

print(water_container(height))
```
