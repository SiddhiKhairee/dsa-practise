 It is used to take down the complexity from O(n^2) to O(n)

Sliding Window = optimize nested loops

Use when:

subarray/substring
contiguous sequence
longest/shortest/maximum/minimum
“at most k”

How to Identify Sliding Window Problems?
These problems generally require Finding Maximum/Minimum Subarray, Substrings which satisfy some specific condition.

The size of the subarray or substring ‘k’ will be given in some of the problems.

These problems can easily be solved in O(n2) time complexity using nested loops, using sliding window we can solve these in O(n) Time Complexity.

Required Time Complexity: O(n) or O(n log n)

### Maximum Average Subarray
You are given an integer array nums consisting of n elements, and an integer k.

Find a contiguous subarray whose length is equal to k that has the maximum average value and return this value. Any answer with a calculation error less than 10-5 will be accepted.



```python
from typing import List
```

This is a brute force method

```python
class Solution:    
    def findMaxAverage(self, nums: List[int], k: int) -> float:

        lp = 0
        rp = (k-1)
        sums = []
        while rp < len(nums):
            sum = 0
            for i in range(lp, rp+1):
                sum += nums[i]
            sums.append(sum)
            lp += 1
            rp += 1

        op = max(sums)/k 
        return op

```

```python
sol = Solution()

# 3. Define the arguments you want to pass
my_nums = [1, 12, -5, -6, 50, 3]
my_k = 4

# 4. Call the function using the instance and store the result
result = sol.findMaxAverage(my_nums, my_k)

# Print the result
print(result)  # Output will be 12.75
```

This is the sliding window technique

```python
class Solution:    
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        ws = 0
        for i in range (0, k):
            ws += nums[i]      # this is the first window sum
        max_sum = ws

        for i in range (k, len(nums)):
            #take the window sum, remove the first element, and add the next element
            ws += nums[i] - nums[i-k]
            max_sum = max(ws, max_sum)

        op = max_sum / k
        return op
```

### Contains Duplicate

```python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        window = set()
        lp = 0
        for rp in range(len(nums)):
            # Check if window has the element or not
            if nums[rp] in window:
                return True
            
            # We need to add elements to the window
            window.add(nums[rp])

            # For all the added elements check, if window size does not exceed k
            if rp - lp >= k:
                # remove leftmost element and shift the left pointer
                window.remove(nums[lp])
                lp += 1
            
        return False
```

```python
sol = Solution()

# 3. Define the arguments you want to pass
my_nums = [1,2,3,1,2,3]
my_k = 2

# 4. Call the function using the instance and store the result
result = sol.containsNearbyDuplicate(my_nums, my_k)

# Print the result
print(result)  # Output will be 12.75
```

```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        smallest = float('inf')
        lp = 0
        w = []
        total = 0
        for rp in range(len(nums)):
            w.append(nums[rp])
            total += nums[rp]
            while total >= target:
                smallest = min(smallest, len(w))
                w.pop(0)
                total -= nums[lp]
                lp += 1

        if(smallest == float('inf')):
            return 0
        else:
            return smallest

```

```python
sol = Solution()


my_nums = [2,3,1,2,4,3]
my_target = 7


result = sol.minSubArrayLen(my_target, my_nums)


print(result)  
```
