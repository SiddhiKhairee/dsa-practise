### Basics of Hashing

```python
# Defining a hashmap
hsmp = {
    1: "yippie",
    2: "I got this",
    3: "I am da goat",
    4: "lock in twin"
}
```

```python
print(hsmp)
```

```python
# accessing an element from hash map
print(hsmp[1])
```

```python
#update the hashmap value
hsmp[1] = "bad bih"
print(hsmp)
```

```python
# insert new value into hashmap
hsmp[4] = "if not you then who"
print(hsmp)
```

```python
if 3 in hsmp:
    print(hsmp[3])
```
