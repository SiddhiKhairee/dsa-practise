### Problem: Frequency Counter

Given a list of integers nums, count how many times each number appears and return the frequencies in a hash map (dictionary).
Input:

nums = [1, 2, 1, 3, 2, 1]

Output:

{
    1: 3,
    2: 2,
    3: 1
}

Explanation:

1 appears 3 times
2 appears 2 times
3 appears 1 time



```python
from typing import List
```

```python
class Solution:
    def freqCounter(self, nums: List[int]) -> dict:
        hsmp = {}
        # count = 1
        for i in range(len(nums)):
            if nums[i] in hsmp:
                hsmp[nums[i]] += 1
            else:
                hsmp[nums[i]] = 1
        return hsmp
```

```python
sol = Solution()


my_nums = [1,1,1,1,1,1]

result = sol.freqCounter(my_nums)


print(result)  
```

### Problem: Character Frequency Counter
Given a string s, count how many times each character appears in the string and return the frequencies in a hash map (dictionary).

Example 1

Input:

s = "banana"

Output:

{
    'b': 1,
    'a': 3,
    'n': 2
}

Explanation:

'b' appears 1 time
'a' appears 3 times
'n' appears 2 times
Example 2

Input:

s = "hello"

Output:

{
    'h': 1,
    'e': 1,
    'l': 2,
    'o': 1
}
Example 3

Input:

s = "aaa"

Output:

{
    'a': 3
}
Function Signature
def charCounter(s: str) -> dict:
    pass
Constraints
1 <= len(s) <= 10^5
s contains English letters only (you can later extend it to spaces, digits, punctuation, etc.)

```python
s = 'banana'
for i in range(len(s)):
    print(s[i])
```

```python
class Solution:
    def freqCounterCharacters(self, s: str) -> dict:
        hsmp = {}
        for c in range(len(s)):
            if s[c] in hsmp:
                hsmp[s[c]] += 1
            else:
                hsmp[s[c]] = 1

        return hsmp
```

```python
sol = Solution()
stringyyy = 'aaaaabaaa'
result = sol.freqCounterCharacters(stringyyy)

print(result)  
```

### Two Sum using Hashmap
Problem: Two Sum

Given an array of integers nums and an integer target, return the indices of the two numbers such that they add up to target.

You may assume that:

Exactly one valid answer exists.
You may not use the same element twice.
The order of the returned indices does not matter.
Example 1

Input:

nums = [2, 7, 11, 15]
target = 9

Output:

[0, 1]

Explanation:

nums[0] + nums[1] = 2 + 7 = 9
Example 2

Input:

nums = [3, 2, 4]
target = 6

Output:

[1, 2]

Explanation:

nums[1] + nums[2] = 2 + 4 = 6
Example 3

Input:

nums = [3, 3]
target = 6

Output:

[0, 1]

Explanation:

nums[0] + nums[1] = 3 + 3 = 6
Function Signature
def twoSum(nums: List[int], target: int) -> List[int]:
    pass
Constraints
2 <= len(nums) <= 10^4
-10^9 <= nums[i] <= 10^9
-10^9 <= target <= 10^9
Exactly one valid answer exists.

```python
a = [12, 34, 56, 78, 90]
if 78 in a:
    print(a.index(78))
```

```python
class Solution:    
    def twoSumUsingHash(self, nums: List[int], target: int) -> List[int]:
        hsmp = {}
        indices = []

        #hashmap will have array value as key nums[i]
        # and the corresponding index as the value i
       
        for i in range (len(nums)):


            # before inserting in hsmp we calc the complement
            complement = target - nums[i]

            # if complement is in the hashmap, means our hashmap key (aka array value) + complement add up to make the target
            # thus, we can return the hashmap value (index of the array where this element is) and we can return i 
            # as i is the index where we found the array element that generated the complement
            if complement in hsmp:
                indices.append(hsmp[complement])
                indices.append(i)

            # simple process of inserting values into hashmap
            # done later so as to avoid duplicate values
            # key -> array element and value -> element's index    
            hsmp[nums[i]] = i

        return indices

    
```

```python
sol = Solution()
my_nums = [2,7,9,20,10]
my_target = 17
result = sol.twoSumUsingHash(my_nums, my_target)

print(result)  
```
